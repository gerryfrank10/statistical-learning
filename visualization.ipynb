```python
import numpy as np
import matplotlib.pyplot as plt
from torch import layout
```

```python
# Line chart
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title("Sine Function")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# Column chart
x = ['A', 'B', 'C', 'D']
y = [10, 15, 12, 29]
plt.bar(x, y)
plt.title("Bar Chart")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# Histogram
data = np.random.randn(1000)
plt.hist(data, bins=30)
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Count")
plt.show()
```

```python
# Bar Chart
x = ['A', 'B', 'C', 'D']
y = [10, 15, 7, 12]
plt.barh(x, y)
plt.title('Bar Chart')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.show()
```

```python
# Stacked Column chart
x = ['A', 'B', 'C', 'D']
y1 = [10, 15, 7, 12]
y2 = [5, 8, 10, 6]
plt.bar(x, y1)
plt.bar(x, y2, bottom=y1)
plt.title('Bar Chart')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.legend(['Y1', 'Y2'])
plt.show()
```

```python
# Pie chart
labels = ['A', 'B', 'C', 'D']
sizes = [30, 25, 15, 30]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.legend(['A', 'B', 'C', 'D'], loc='upper left')
plt.show()
```

```python
# Donut Chart
sizes = [30, 25, 15, 30]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', wedgeprops={'edgecolor':'white'})
plt.title('Donut Chart')
plt.legend(['A', 'B', 'C', 'D'], loc='upper left')
plt.show()
```

```python
# Area chart
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.fill_between(x, 0, y, alpha=0.7)
plt.title('Area Chart')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Scatter plot
x = np.random.randn(100)
y = np.random.randn(100)
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Box plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data)
plt.title('Box Plot')
plt.xlabel('Data')
plt.ylabel('Value')
plt.show()
```

```python
# KDE chart Using seaborn
import seaborn as sns
data = np.random.randn(1000)
sns.histplot(data, bins=30, kde=True)
plt.title('Histogram with KDE')
plt.xlabel('Values')
plt.ylabel('Density')
plt.show()
```

```python
# KDE using Gaussian KDE
import numpy as np
import scipy

data = np.random.randn(1000)
plt.hist(data, bins=30, density=True)
plt.title('Histogram with density')
plt.xlabel('Values')
plt.ylabel('Density')

kde = scipy.stats.gaussian_kde(data)
x = np.linspace(data.min(), data.max(), 100)
plt.plot(x, kde(x))
plt.legend(['KDE'])
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = ['A', 'B', 'C', 'D']
y = [10, 15, 12, 29]
y2 = [1, 10, 22, 10]
bar_width = 0.35

x1_positions = range(len(x))
x2_positions = [p + bar_width for p in x1_positions]

plt.bar(x1_positions, y, width=bar_width, label='y1', color='blue', alpha=0.7)
plt.bar(x2_positions, y2, width=bar_width, label='y2', color='red', alpha=0.7)
plt.title('Bar Chart')
plt.xlabel('Values')
plt.ylabel('Count')
plt.legend(['y1', 'y2'], loc='upper left')
plt.show()
```

```python
# Or
import matplotlib.pyplot as plt
import numpy as np

species = ("Adelie", "Chinstrap", "Gentoo")
penguin_means = {
    'Bill Depth': (18.35, 18.43, 14.98),
    'Bill Length': (38.79, 48.83, 47.50),
    'Flipper Length': (189.95, 195.82, 217.19),
}
x = np.arange(len(species))
width = 0.25
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

x2 = [p + bar_width for p in x]
for key, value in penguin_means.items():
    offset = multiplier * width
    rects = ax.bar(x + offset, value, width=width, label=key)
    ax.bar_label(rects, padding=3)
    multiplier += 1

ax.set_xticks(x + width)
ax.set_xticklabels(species)
ax.legend(loc='upper left', ncols=3)
plt.show()
```
