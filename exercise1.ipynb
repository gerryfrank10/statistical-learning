```python
# f. Create a new qualitative variable ELITE, by binning Top10perc
df['Elite'] = pd.cut(df['Top10perc'],[0, 50, 100], labels=['No', 'Yes'])
print(df['Elite'].value_counts())
```

```python
df.boxplot(column=['Outstate'], by='Elite')
```
